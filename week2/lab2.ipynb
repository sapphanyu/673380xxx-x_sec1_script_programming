Lab 2.1: Number Classifier
○ Objective: Students will write a script to classify a user-provided integer as positive,
negative, or zero, and also check if it's even or odd.

```python
num = int(input("Enter a number: ") )
if num < 0:
    text = f"{num} is negative"
elif num > 0:
    text = f"{num} is positive"
else:
    text = f"{num} is zero"

if num == 0:
    print(f"{text}.")
elif num % 2 == 0:
    print(f"{text} and even number.")
else:
    print(f"{text} and odd number.")
```

Lab 2.2: Simple Decision-Making Program (Movie Rating)
○ Objective: Students will create a program that recommends a movie rating based on
the user&#39;s age, demonstrating nested or chained conditionals.
○ Tasks (Common for Local &amp; Colab):
1. Create a new Python file (lab2_2.py) or Colab code cell.
2. Prompt the user to enter their age. Convert it to an integer.
3. Implement the following logic:
■ If age is less than 5: Print &quot;You&#39;re too young for movies! Enjoy cartoons.&quot;
■ If age is 5 to 12 (inclusive): Recommend &quot;G-rated or PG-rated movies.&quot;
■ If age is 13 to 17 (inclusive): Recommend &quot;PG-13 or R-rated (with parental
guidance).&quot;
■ If age is 18 or older: Recommend &quot;Any movie rating.&quot;
4. Challenge: Add a secondary prompt: &quot;Do you like action movies? (yes/no)&quot;. If the
age is 18 or older AND they like action, print &quot;You might enjoy the latest action
blockbuster!&quot; (This requires a nested if or more complex logical and conditions).

```python
age = int(input("Enter your age: "))
if age < 5:
    print("You're too young for movies! Enjoy cartoons.")
elif age <= 12:
    print("You can watch G-rated movies.")
elif age <= 17:
    print("You can watch PG-13 movies.")
else:
    print("You can watch anything rated movies.")
```
